# The goal of the project: to measure the extent of understanding of the concepts of the numpy library

```python
import numpy as np
```

## Create an array of 15 floats numbers using linspace

```python
arr = np.linspace(0,2,15)
print(arr)
```

## Create a three-dimensional (3D) matrix in the form (2, 2, 3)

```python
my_3d_arr = np.array([[
    [1.1,2.1,3.1], [4.1,5.1,6.1]]
    ,[[7.1,8.1,9.1],[10.1,11.1,12.1]]])
print(my_3d_arr)
print("shape of array is: ",my_3d_arr.shape)
```

## Print the properties of the array

```python
print("Type : ",type(my_3d_arr))
print("Data type : ",my_3d_arr.dtype)
print("Dimensions : ",my_3d_arr.ndim)
print("Shape : ",my_3d_arr.shape)
print("Size : ",my_3d_arr.size)
```

## Modify the array to become 4dimensions (4D)

```python
my_4d_arr = np.array(my_3d_arr,ndmin=4)
print("New array :\n",my_4d_arr)
print("New dimensions : ",my_4d_arr.ndim)
print("New shape : ",my_4d_arr.shape)
```

## Change the array type to integer

```python
new_arr = my_4d_arr.astype(int)
print(new_arr)
print("New data type: ",new_arr.dtype)
```

## Print all element of the array using nditer()

```python
for x in np.nditer(new_arr):
  print(x , end=" ")
  print()
```

## Print the 8th elements in the array

```python
print("8th element: ",new_arr[0,1,0,1])
```

## Print the element number 5 and 6 in the array

```python
print("5th and 6th elements: ",new_arr[0,0,1,1:])
```

## Search for the element number 8 using where ()

```python
print(np.where(new_arr==8))
```

## Reshape the array to (2,3,2)

```python
re_new_arr = new_arr.reshape(2,3,2)
print(re_new_arr)
```

## Merge the two array without specifying axis

arr1 = np.array([['A', 'B'], ['E', 'F']])
arr2 = np.array([['C', 'D'], ['G', 'H']])

```python
arr1 = np.array([['A', 'B'], ['E', 'F']])
arr2 = np.array([['C', 'D'], ['G', 'H']])

joined_arr = np.concatenate((arr1,arr2))
print(joined_arr)
```

## Merge two arrays based on rows axis = 1

```python
joined_arr2 = np.concatenate((arr1,arr2),axis=1)
print(joined_arr2)
```

## Spilt the array into two arrays with axis=1

```python
spilt_joined_arr2 = np.array_split(joined_arr2,2,axis=1)
print(spilt_joined_arr2)
```
